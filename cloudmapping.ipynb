```python
allpower0=[]
allpower1=[]
allpcal0=[]
allpcal1=[]
alll=[]
allb=[]
for i in range(len(files)):
    a=np.load(files[i])
    power0=a["power0"]
    power1=a["power1"]
    pcal0=a["pcal0"]
    pcal1=a["pcal1"]
    l=a["l"]
    b=a["b"]
    allpower0.append(power0)
    allpower1.append(power1)
    allpcal0.append(pcal0)
    allpcal1.append(pcal1)
    alll.append(l)
    allb.append(b)
allpower0=np.array(allpower0)
allpower1=np.array(allpower1)
allpcal0=np.array(allpcal0)
allpcal1=np.array(allpcal1)
alll=np.array(alll)
allb=np.array(allb)
a=np.load(files[0])
print(a.keys())
```

```python
for i in range(31):
    plt.plot(allpower0[i])
plt.show()
```

```python
off=allpower0[0]
```

```python
alldivspec=[]
for spectrum in allpower0:
    divspectrum=spectrum/off
    alldivspec.append(divspectrum)
alldivspec=np.array(alldivspec)
```

```python
for i in range(31):
    plt.plot(alldivspec[i])
plt.show()
```

```python
plt.plot(alldivspec[0])
plt.show()
```

```python
print(np.argmax(alldivspec[0]))
```

```python
adbs=[]

for spec in alldivspec:
    print(np.argmax(spec))
    spec[673:677]=1
    spec-=np.median(spec)
    adbs.append(spec)
```

```python
for i in range(31):
    plt.plot(adbs[i])
plt.show()
```

```python
tvec=[]
for spec in adbs:
    t=np.max(spec)
    tvec.append(t)
```

```python
alll_m = []
for a in alll*np.pi/180:
    if a >= np.pi:
        new_a = a - 2*np.pi
    else:
        new_a = a
    alll_m.append(new_a)
```

```python
plt.subplot(projection="mollweide")
plt.scatter(alll_m,allb*np.pi/180,c=tvec)
plt.grid()
plt.show()
```
